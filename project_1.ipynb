```python
dp = pd.read_csv('시군별 치매환자.csv')
dp = dp[['지역','비율']]
dp['비율'] = pd.to_numeric(dp['비율'].values)
dp.columns = ['지역구','비율']
dp
```

```python
import csv

def convert_to_float(input_file, output_file):
    with open(input_file, 'r', newline='') as csvfile:
        reader = csv.reader(csvfile)
        data = list(reader)

    for i in range(1,len(data)):
        for j in range(1,len(data[i])):
            try:
                data[i][j] = float(data[i][j])
            except ValueError:
                # If the conversion to float fails (e.g., non-numeric data), you can handle it here.
                # For simplicity, we'll just leave the original value as is.
                pass

    with open(output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows(data)

# Usage example:
input_file_path = 'heatmap.csv'
output_file_path = 'HM.csv'
convert_to_float(input_file_path, output_file_path)
```

```python
satisfaction_df = satisfaction[(satisfaction["전반적인 여가활동 매우 만족 (퍼센트)"].isnull() == False) & (satisfaction["문화여가시설(접근성과 충분 정도) 5점 평균 (점)"].isnull() == False)]
satisfaction_df
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()    #model 생성
model.fit(satisfaction_df[["문화여가시설(접근성과 충분 정도) 5점 평균 (점)"]], satisfaction_df[["전반적인 여가활동 5점 평균 (점)"]])   
```

```python
r_sq = model.score(satisfaction_df[["문화여가시설(접근성과 충분 정도) 5점 평균 (점)"]], satisfaction_df[["전반적인 여가활동 5점 평균 (점)"]])
 
print(r_sq)
print(model.intercept_)
print(model.coef_)
```

```python
satisfaction_df["predicted"] = model.predict(satisfaction_df[["문화여가시설(접근성과 충분 정도) 5점 평균 (점)"]])
 
sns.relplot(x= "문화여가시설(접근성과 충분 정도) 5점 평균 (점)", y = "전반적인 여가활동 5점 평균 (점)", data = satisfaction_df)
sns.lineplot(x= "문화여가시설(접근성과 충분 정도) 5점 평균 (점)", y = "predicted", color="r", data = satisfaction_df)
```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(place_juso):
    try:
        geo = geo_local.geocode(place_juso)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
latitude = []
longitude =[]

for i in place_juso:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```
